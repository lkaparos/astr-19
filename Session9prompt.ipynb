```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
n=1000
x=np.random.uniform(0,1,1000)
```

```python
plt.hist(x, bins=100, color="coral", edgecolor="black")
width=0.5
histmin=np.floor(min(x))
histmax=np.ceil(max(x))+width
plt.xlabel('x')
plt.ylabel('n per bin')
```
